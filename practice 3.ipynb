```python
import pandas as pd
df = pd.read_json("/Users/alcedreicbangcola/Downloads/restaurant_orders_with_price_and_date.json")
df
```

```python
unique_customers = set(df['customer'])
print(len(unique_customers))
```

```python
unique_products  = set(df['product_name'])
print(unique_products)
print(len(unique_products))
```

```python
df['times_ordered'] = df.groupby('order_id')['customer'].transform("count")
most_orders = df.nlargest(1, 'times_ordered')
name_of_suki = most_orders['customer']
print(name_of_suki)
```

```python
df['total_sales'] = df['unit_price']*df['quantity']*df['times_ordered']
product_revenue = df.groupby("product_name")["total_sales"].sum().reset_index()

# Step 3: Get the top 3 products by revenue
top3_products = product_revenue.nlargest(3, 'total_sales')

print(top3_products)
product_revenue
```

```python
#Instructions

#Convert each transaction into individual line items. Each row should contain only one product.
#Extract the SKU (Stock Keeping Unit) by keeping the product name without the quantity and create a 'SKU' column.
#Extract the quantity without the product name and create a 'quantity' column.
#Extract the package size (g, kg, ml, L) and create a 'package_size' column. If the SKU does not have a package size, print "NA".
#Count line items per order and create a line_item_count column that displays the total number of products per order.
```

```python
import pandas as pd

canteen_order = pd.read_json("/Users/alcedreicbangcola/Downloads/canteen_orders.json")

canteen_order.head()
```

```python
canteen_order['items'] = canteen_order['items'].astype(str)
canteen_order['items'] = canteen_order['items'].str.split(';')

canteen_order = canteen_order.explode('items').reset_index(drop=True)


canteen_order.head()
```

```python
canteen_order['SKU'] = canteen_order['items'].str[:-5]
canteen_order.head()
```

```python
canteen_order['Quantity'] = canteen_order['items'].str[-2:-1]
canteen_order.head()
```

```python
unit_sizes = ['ml','g','L',]
package_sizes = []
for item in canteen_order['SKU']:
    size = "NA"
    for unit in unit_sizes:
        if item.endswith(unit):
            size = item.split()[-1] 
    package_sizes.append(size)
canteen_order["Size"] = package_sizes
canteen_order.head()

            
            
   
    
```

```python
canteen_order["Line Count"] = canteen_order.groupby('customer_id')['SKU'].transform("count")
canteen_order.head(10)

```

```python
# Instructions

# You are given a dataset of bookstore transactions. Each row represents an order that may contain multiple books in a single string.

# Your tasks:

# Convert each transaction into individual line items. Each row should contain only one book.

# Extract the Book Title by removing the quantity and format, and create a book_title column.

# Extract the Quantity (number before each book) and create a quantity column.

# Extract the Format (hardcover, paperback, ebook) and create a format column. If the book does not have a format, return "NA".

# Count line items per order and create a line_item_count column that shows the total number of books per order.
```

```python
import pandas as pd

df = pd.read_csv("/Users/alcedreicbangcola/Downloads/book_orders.csv")  

df
```

```python
df['products'] = df['products'].astype(str)
df['products'] = df['products'].str.split(',')
df = df.explode('products')
df
```

```python
df['Book Title'] = df['products'].str[2:]
df
```

```python
df['Quantity'] = df['products'].str[0:2]
df
```

```python
import re

material = []
for product in df['products']:
    match = re.search(r"\((.*?)\)", product)  #  r"\((.*?)\)",  "Find text inside the first pair of parentheses."
    # r"\((.*?)\)" This looks for text inside parentheses (...)
    if match:
        material.append(match.group(1))  
    else:
        material.append("NA")             

df['Format'] = material
df
```

```python
df['Line Item Count'] = df.groupby('order_id')['Book Title'].transform("count")
df
```
